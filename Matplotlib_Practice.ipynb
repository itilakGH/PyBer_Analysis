```python
#The % sign is sometimes called the magic command. When you run a cell that starts with %, 
#“magic” happens behind the scenes and sets the back-end processor used for charts. 
#Using it with the inline value displays the chart within the cell for a Jupyter Notebook file.

%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis_1 = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
y_axis_2 = [9.02, 20.24, 40.20, 32.42, 30.34, 33.04, 45.82, 15.56, 11.85, 25.90, 23.71, 21.09]
y_axis_3 = [10.72, 25.24, 37.20, 35.02, 33.34, 28.04, 41.82, 15.56, 13.85, 29.90, 25.71, 21.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis_1)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis_1)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis_1)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis_1)
```

```python
# Create the plot with plt.show().
plt.plot(x_axis, y_axis_1)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis_1, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis_1, marker="*", color="green", linewidth=2, label='Boston')
plt.plot(x_axis, y_axis_2, marker="*", color="yellow", linewidth=2, label='New York')
plt.plot(x_axis, y_axis_3, marker="*", color="Blue", linewidth=2, label='San Francisco')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Making changes to the chart with the Object-Oriented Interface
# Create the plot with ax.plt()
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x_axis, y_axis_1, label='Boston', color="green", linewidth=3, marker='D')
ax.plot(x_axis, y_axis_2, label='NY', color="blue", linewidth=3, marker='D')
ax.plot(x_axis, y_axis_3, label='SF', color="red", linewidth=3, marker='D')
ax.set_ylim(0, 47)
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
ax.grid()
```

```python
#Creating a dataframe
import pandas as pd
index = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
Boston = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
New_York = [9.02, 20.24, 40.20, 32.42, 30.34, 33.04, 45.82, 15.56, 11.85, 25.90, 23.71, 21.09]
San_francisco = [10.72, 25.24, 37.20, 35.02, 33.34, 28.04, 41.82, 15.56, 13.85, 29.90, 25.71, 21.09]
ride_rates_df = pd.DataFrame({'Boston': Boston, 'New York': New_York, 'San Francisco': San_francisco}, index=index)
ax = ride_rates_df.plot.bar(rot=0, width=0.7, figsize=(20,10))
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis_1, label='Boston', color="green")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot (horizontal bar chart)
plt.barh(x_axis, y_axis_1, label='Boston', color="green")

# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot (horizontal bar chart)
plt.barh(x_axis, y_axis_1, label='Boston', color="magenta")
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot (horizontal bar chart) with OOI
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis_1, label='Boston', color="k")
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend()
ax.grid()
```

```python
school_name = ["Hernandez High School", "Figueroa High School", "Wilson High School","Wright High School",
               "Thomas High School","Fernando High School","Eagles High School"]
budget = [15773, 29408, 21319, 18005, 25100, 25500, 23007]
plt.pie(budget, labels=school_name)
plt.show()
```

```python
explode_values = (0,0,0,.2,0,0,0)
colors = ["mistyrose", "lightcoral", "rosybrown", "plum", "lightpink", "thistle", "wheat"]
plt.pie(budget, explode=explode_values, labels=school_name, autopct='%.1f', colors=colors)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

explode_values = (0,0,0,0,0,0,.2)
colors = ["mistyrose", "lightcoral", "rosybrown", "plum", "lightpink", "thistle", "wheat"]
ax.pie(budget, labels=school_name, explode=explode_values, autopct='%.1f%%', colors=colors)
ax.set_title("PyBer Fare by Month")
```
